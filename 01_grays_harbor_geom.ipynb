This notebook 
- reads in counties geodataframe from census.gov
- filters Grays Harbor County
- extracts Geoseries
- converts to crs 26910 to match lidar DEM
- isolates the geometry attribute
- saves geodataframe 

```python
import geopandas as gpd
```

```python
# https://gda-wi26-jupyterbook.readthedocs.io/en/latest/modules/06_Raster2/06_demo_download.html

url = "https://www2.census.gov/geo/tiger/TIGER2023/COUNTY/tl_2023_us_county.zip"
counties = gpd.read_file(url)
grays_harbor = counties[counties["NAME"] == "Grays Harbor"]
grays_harbor_26910 = grays_harbor.to_crs("EPSG:26910")
grays_harbor_geom = grays_harbor_26910.geometry.squeeze()
```

```python
grays_harbor_geom
```

```python
grays_harbor_26910.to_file("../processed_data/grays_harbor.geojson", driver="GeoJSON")
```
